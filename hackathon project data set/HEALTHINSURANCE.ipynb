```python
import pandas as pd
path="C:\\Users\sunid\\Desktop\\hackathon project data set\\insurance.csv"
data=pd.read_csv(path)
data
```

```python
data.info()
```

```python
print(data['sex'].unique())
print(data['smoker'].unique())
print(data['region'].unique())
```

```python
data['sex']=data['sex'].map({'female':0,'male':1})
data['smoker']=data['smoker'].map({'yes':0,'no':1})
data['region']=data['region'].map({'southwest':0,'southeast':1,'northwest':2,'northeast':3})
data
```

```python
inputs=data.drop('charges',axis=1)
outputs=data.drop(['age','sex','bmi','children','smoker','region'],axis=1)
print(inputs)
print(outputs)
```

```python
import sklearn 
from sklearn import linear_model
model=linear_model.LinearRegression() 
model.fit(inputs,outputs)

a=model.predict([[32,1,28.880,0,1,2]])
print(a[0])

acc=model.score(inputs,outputs)
print("the accuracy for the given data will be:",int(acc*100),"%")
```
